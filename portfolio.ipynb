```python
import yfinance as yf
stocks = [ 'AAPL', 'MSFT', 'TSLA']
data = yf.download(stocks, start='2020-01-01')['Close']
data.head()

print('calculate the daily rerturns')

daily_return = data.pct_change().dropna()
print(daily_return.head())

print('portfolio returns')
weights = [1/len(daily_return.columns)]*len(daily_return.columns)
port_return = daily_return.dot(weights)
cum_return = (1+port_return).cumprod()

print(cum_return.head())
cum_return.plot()

avg_return = port_return.mean()
print("this is average return", avg_return)
volatility = port_return.std()
print("this is volatiltiy", volatility)
sharpe_ratio = avg_return/volatility
print("sharpe ratio:" ,sharpe_ratio)
```
